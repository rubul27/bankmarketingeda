```python
#lets see correlation of all variables wrt deposits
# With that we can determine if duration has influence on term deposits

from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
# fig = plt.figure(figsize=(12,8))
df['deposits'] = LabelEncoder().fit_transform(df['deposits'])



# Separate both dataframes into numeric
numeric_df = df.select_dtypes(exclude="object")
# categorical_df = df.select_dtypes(include="object")

corr_numeric = numeric_df.corr()


sns.heatmap(corr_numeric, cbar=True, cmap = 'coolwarm')  #, cmap="RdBu_r"
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

```python
corr_numeric
```

```python
# we see that deposits are very highly correlated with durations
```

```python
avg_duration = df['duration'].mean()
print(avg_duration)
df["duration_status"] = np.nan

for ind, row in df.iterrows():
  if row['duration'] >= avg_duration:
    df.loc[ind, 'duration_status'] = 'above_average'
  else:
    df.loc[ind , 'duration_status'] = 'below_average'

pct_term = pd.crosstab(df['duration_status'], df['deposits']).apply(lambda r: round(r/r.sum(), 2) * 100, axis=1)

pct_term

```

```python
ax = pct_term.plot(kind='bar', stacked=False, cmap='RdBu')
plt.title("The Impact of Duration \n in Opening a Term Deposit", fontsize=18)
plt.xlabel("Duration Status", fontsize=18);
plt.ylabel("Percentage (%)", fontsize=18)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.02, p.get_height() * 1.02))


plt.show()
```
